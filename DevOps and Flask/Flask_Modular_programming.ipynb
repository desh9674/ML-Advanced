##Flask Command Line Interface
In Flask environment, a script named flask is available which provides commands from Flask, extensions, and the application.

When executed from a terminal, it displays the usage syntax of flask script, various options and commands available with it.

Syntax of flask script
```
flask [OPTIONS] COMMAND [ARGS]
```
When flask script is executed with any of the allowed commands, it loads the application defined in **FLASK_APP** environment variable or **app.py** file or **wsgi.py** file.

#Jinja's Syntax
Jinja's syntax is used while writing templates.

In jinja, the value of a variable can be accessed by placing it in between two open braces and closed braces, i.e., {{ }}. It is known as variable block.

For example, the expression {{ x }} is substituted by a value of variable x, when the template is rendered.

Similarly, there are control blocks defined by {% %}. They are used to specify if blocks, forblocks, and customized blocks.



#Variables Substitution
Any Python object can be passed as a variable to jinja template and can be used in variable substitution block, as long as the object can be represented as a string using str function.

Assume a list, ```k = [8, 15, 25, 88, 12, 78]```, a dictionary, ```d = {'x':'loyd', 'y':'clive', 'z':'patti'}``` and an object o having a property city with value Chicago.

The following expressions show how to access elements of a list k, dictionary d and city attribute of object o.
```
{{ k[2] }} - returns 25.

{{ d['y'] }} - returns clive.

{{ o.city }} - returns Chicago.
```

#Filters
Filters are the built-in or defined functions, which alter the value of variables for display purpose.

These functions cannot be used directly in templates.

Instead, in jinja a filter is used in a variable block with a pipe character.

Syntax of using a filter
```
{{ variable | filter_name(*args) }}
```
If the filter doesn't require any arguments, then the syntax looks as below.
```
{{ variable | filter_name }}
```
In the upcoming pages, you will see some of the filters used in templates.


#Filter - default
If any variable holds None value, then it can be replaced with a default value using default filter.

In the below example, variable a holds None value. Hence, it displays Default Value message to user when the template is rendered.

Usage of default
```
{{ a | default('Default Value') }}
```
Output
Default Value
You can also use default filter to replace a variable value if it is evaluated to be False. In such a case, True needs to be passed as the second parameter to default filter as shown below.
```
{{ '' | default('Default Value', True) }}
```


#Filter - int, float, round
int, float filters convert passed a variable value into integer and float respectively.
```
Usage of int
{{ 67.7 | int }}  
Output
67
Usage of float
{{ 75 | float }}
Output
75.0
round filter rounds off a float value to a specified number of decimals.
Usage of round
{{ 5.16725 | round }}       ----> 5.0
{{ 5.16725 | round(2) }}    ----> 5.17
{{ 5.16725 | round(3) }}    ----> 5.167
{{ 5.7 | round(0, "floor") }} --> 5.0
{{ 5.2 | round(0, "ceil") }}  --> 6.0

```
The output of each jinja expression is shown on same line above.


#Filter - join
join filter is used to combine all elements of a list, specified in a variable block.

Each element of the list is then converted to a string before joining each other.

The below example illustrates how join filter works when a list ```k = [8, 15, 25, 88, 12, 78]``` is passed to a template in which below jinja expression is written.
```
Usage of join
{{ k | join(',') }}    ---> 8,15,25,88,12,78 
{{ k | join('-') }}    ---> 8-15-25-88-12-78
```

#Filter - length
length filter is used to return the length of the variable.

The below example shows use of length filter, which determines the length of the list k = [8, 15, 25, 88, 12, 78], passed to a template.
```
Usage of length
{{ k | length }}
Output
6
```

#Chaining Filters
Filters can be chained together.

The below-shown example initially converts all characters of string s into the lower case with lower filter.

The lowercase version of string s is further converted the into Title case using title filter.
```
{{ s | lower | title }}
```
When a template with the above expression is rendered, it displays the following output to the user.
```
Hello World
```


Custom Filters
In addition to built-in filters, it is also possible to define custom filters and use them in templates.

A custom filter is a normal Python function. It has to be defined initially and then added to the application environment variable jinja_env.filters.

The below example defines a custom filter startswith in hello.py file.
```
Custom Filter
def startswith(word, tag):
    if word.startswith(tag):
        return tag + '-word'

    return 'non-' + tag + '-word'

app.jinja_env.filters['startswith'] = startswith

```


#Using a Custom Filter
The custom filter startswith defined in a previous slide, checks if a string starts with a specific tag or not.

If the string starts with specified tag, it returns <tag>-word, else it returns non-<tag>-word.

Usage of startswith
Example 1
```
{{ 'university' | startswith('un') }}
Output
un-word
Example 2
{{ 'university' | startswith('pre') }}
Output
non-pre-word
```

#One-to-Many Relationship

#Class Based Views
Till now, you have worked with views, written as functions.

However, in case if many view functions share similar functionality, then it is advisable to implement views as classes.

In this topic, you will learn how to write Class Based Views.

#Defining a Class-Based View
The most frequent activity that you do would be to render a template.

The below example defines a BaseView class, which accepts a template_name as input and capable of rendering the template, whose functionality is defined under dispatch_request method.

Add below example to hello.py
```
hello.py
from flask.views import View

class BaseView(View):
    def __init__(self, template_name):
        self.template = template_name
        super(BaseView, self).__init__()

    def dispatch_request(self):
        return render_template(self.template)
```


#Adding a Rule
After defining BaseView class, a rule can be added using app.add_url_rule function.

For example, now let's add the below statement to hello.py. It adds /users/ rule, which renders listusers.html template.
```
app.add_url_rule('/users/', view_func=BaseView.as_view('display_users', template_name='listusers.html'))
```

The function app.add_url_rule links a route to a function call.

The first argument defines the function's route, and view_func argument takes the view function, as value, that handles the route.

The expression BaseView.as_view transforms BaseView into a view function.

Name of the view function is passed as the first argument to as_view function. The remaining parameters passed to as_view are passed to __init__ method of BaseView class.

class ListView(BaseView):
    def __init__(self, template_name, model):
        self.model = model
        super(ListView, self).__init__(template_name)

    def get_objects(self):  # used to fetch objects in db
        objects = self.model.query.all()
        return {self.model.__tablename__+'s':objects}

    def dispatch_request(self): # used to display fetched objects in html
        return render_template(self.template, **self.get_objects())

# Blueprints

In Flask, Blueprints are used for developing components of an application that support common patterns, within an application or across applications.

A Blueprint object behaves similarly to a Flask application object. However, it is not an application and used just to add/extend the functionality of an existing application.

The above GIF mimics the functionality of a blueprint, by grouping related items together.

#Uses of Blueprints
Blueprints are generally used, in Flask, for the following purposes:

To organize the entire application to individual components.

To make the blueprint of an application available only at a specific URL prefix.

To make the blueprint of an application available multiple times with varied URL rules.


#Your First Blueprint
Let's create a new file sample_blueprint.py and add the below shown content.


sample_blueprint.py

```
from flask import Blueprint, render_template

sampleBP = Blueprint(
    'sample',
    __name__,
    template_folder='templates/sample',
    static_folder='static/sample',
    url_prefix="/sample"
)

@sampleBP.route('/')

    return render_template('home.html')

```

The Blueprint class takes two required arguments. The first one being the name of the blueprint and the second one is the name of the package.

The other passed parameters, such as template_folder, static_folder, are optional. They specify the path where the defined blueprint will look for files.

Argument url_prefix is also optional. The value passed to url_prefix argument is added, automatically, to start of every route defined in the blueprint.

#Application Testing in Flask


As you know, Application Testing is an important step that checks if an application is behaving as expected, based on the written code.

There are mainly three types of testing: Unit Testing, Integration Testing and System Testing.

Unit Testing: The tests that verify the correctness of smallest possible units of code, such as functions.

Integration Testing: It checks if different interacting components work as expected.

System Testing: The tests that check the correctness of the whole system.

Testing Flask Applications
A Flask application comes with a test client, which is capable of simulating HTML requests and returns a response object.

You must write test cases, such that they test every function with data, covering every part of the function.

In this topic, you will get to see how to write test methods for two view functions available in hello.py.

#Deploying a Flask Application
After successful testing of the application, it has to be moved into production.

There are a number of approaches to host a Flask application.

Most of the Python applications are run on WSGI compatible servers.

In addition, Flask applications can be hosted on Platform-as-a-Service (PaaS) platforms like Heroku, Amazon web services, and Google Cloud.

In this topic, you will get a glimpse of deploying Flask application on few standalone WSGI servers and Heroku platform.


#Deploying on 'Gevent'
Gevent is a coroutine-based Python library. It has a simple interface to run WSGI applications.

To deploy your project, create a new file gserver.py in project folder, with the below content.
```
from gevent.pywsgi import WSGIServer

from project import create_app

app = create_app()

server = WSGIServer(('', 80), app)
server.serve_forever()
```

Then, you can then run the server with command python gserver.py.

#Deploying on `Gunicorn`
Gunicorn is a WSGI HTTP server for Unix.

In addition to other Python web servers, it has a good user-interface which is easy to use and configure.

To deploy your project, create a new file wsgi.py in project with below content.
```
from project import create_app

app = create_app()
```

Then you can run the server using the below command.
```
gunicorn -w 4 myproject.wsgi:app
```

The above command runs the Flask application with four worker processes.

#Deploying on uWSGI
uWSGI is used as an application server for many programming languages including Python.

uWSGI can run either as a standalone server or behind as a full web server such as Nginx or Apache.

Similar to the usage of Gunicorn, wsgi.py file with the following content needs to be created in project folder.
```
from project import create_app

app = create_app()

```

Deploying on uWSGI
Now run the server with the below command.
```
uwsgi --socket 127.0.0.1:8080 --wsgi-file wsgi.py --callable app --processes 6 --threads 3
```

Rather than using command line options, you can also specify the options in a new file uwsgi.ini, as shown below and use the command $ uwsgi uwsgi.ini

```
[uwsgi]
socket = 127.0.0.1:8080
wsgi-file = wsgi.py
callable = app
processes = 6
threads = 3
```

#Deploying with Nginx and uWSGI
In case, if more performance of a web application is expected, it is recommended to use a web server such as Nginx as a front end to WSGI server such as uWSGI, by using a reverse proxy.

You need to create ngnix.conf file along with uwsgi.ini, shown previously, and place it in /etc/nginx/sites-available/<directory> folder.

ngnix.conf configurations, similar to one shown below.

```
server {
    listen 80;
    your_server_name your_domain_name;

    location / {
        include uwsgi_params;
        uwsgi_pass 127.0.0.1:8080;
    }
    
    location /static {
        alias /home/deploy/project/static;
    }
}

```


The above configurations tell Ngnix to listen for incoming requests on 80 and redirect all the requests to WSGI application listening on 8080.

It also makes an exception for static file requests.


#Deploying on Heroku
Heroku is a PaaS. An application deployed on Heroku works based on commands written in Procfile.

These commands are run by Heroku dyno, i.e., a virtual machine.

Hence create a file Procfile in the top-level directory of your project and add below content.

```
web: uwsgi uwsgi.ini

```

The above contents indicate that there is one process web, which will run uwsgi command with parameters specified in uwsgi.ini file.

Also, create a file runtime.txt and specify the python runtime environment that you would like to use, as shown below.
