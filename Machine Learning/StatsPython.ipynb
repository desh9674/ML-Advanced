```
import statsmodels.api as sm

biopsy_set = sm.datasets.get_rdataset('biopsy','MASS')

df = biopsy_set.data

#print(df.columns)
#print(df.head(10))
#bio_data = bio_data.rename(index = {"class":"Class"}, inplace = True)
df = df.rename(columns = {'class':'Class'})
#print(df.head(10))
#bio_data2.class.map(dict(bengin=0, malignant=1))
df['Class'] = df['Class'].map({'bengin': 0.0,'malignant': 1})
df['Class'].fillna(0, inplace=True)
#df.Class.replace(to_replace=['bengin', 'malignant'], value=[0, 1])
#print(df.head(10))

import statsmodels.formula.api as smf

logit = smf.logit('Class ~ V1', df)
results = logit.fit()
print(results.prsquared)

```

```
import statsmodels.api as sm

icecream = sm.datasets.get_rdataset("Icecream", "Ecdat")
icecream_data = icecream.data


import statsmodels.formula.api as smf
from scipy import stats

model1 = smf.ols('cons ~ temp', icecream_data).fit()

from statsmodels.stats import anova
print(anova.anova_lm(model1))

model2 = smf.ols('cons ~ income + temp', icecream_data).fit()
print(anova.anova_lm(model2))
print(stats.f.sf(31.81, 2, 27))  #probability computation
print(anova.anova_lm(model1, model2)) # compare two models residual sum squares ssr, if reduced then model is significant
```
