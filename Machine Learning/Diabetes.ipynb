```python
import pickle
import hashlib
def get_pickle(file_name):
    with open(file_name, 'rb') as f:
        return pickle.load(f)
```

```python
def geth(obj):
    obj = str(obj).encode()
    m = hashlib.md5()
    m.update( bytes(obj) )
    #print(m.hexdigest())

    return m.hexdigest()
```

```python
def make_pickle(file_name, obj):
    with open(file_name, 'wb') as f:
        pickle.dump(geth(obj), f, pickle.HIGHEST_PROTOCOL)
```
